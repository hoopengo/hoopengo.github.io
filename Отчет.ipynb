4. Решение задачи 4, где нужно создать ссылки на дочерние странички

1. `<ul>` - Этот тег используется для начала неупорядоченного списка, который представляет собой список элементов, отмеченных маркерами.

2. `<li><a href="./dir1.html">ссылка 1</a></li>` - Эта строка создает элемент списка с гиперссылкой. Атрибут href в теге `<a>` указывает URL страницы, на которую ведет ссылка. В данном случае он ссылается на файл с именем dir1.html в той же директории, что и текущая страница. Текст между тегами `<a>` и `</a>` ("ссылка 1") - это текст, который будет отображаться для ссылки.
Следующие две строки аналогичны, но ссылаются на `dir2.html` и `dir3.html` соответственно.

3. `</ul>` - Этот тег используется для завершения неупорядоченного списка.

6. Решение задачи 6, где нужно разместить фотографии, аудио и видео.

В ходе выполнения задания были предоставлены инструкции по размещению медиа-контента на веб-страницах с использованием HTML.

Фотографии: Для размещения изображений на веб-странице используется тег `<img>`. В атрибуте src указывается путь к изображению, а в alt - альтернативный текст, который будет отображаться в случае, если изображение не может быть загружено.

Пример кода:
`<img src="path_to_your_image.jpg" alt="Some text">`

Аудио: Для вставки аудио на веб-страницу используется тег `<audio>`. Внутри этого тега размещается тег `<source>`, в атрибуте src которого указывается путь к аудиофайлу. Также внутри тега `<audio>` можно разместить текст, который будет отображаться в браузерах, не поддерживающих этот тег.

Пример кода:

```
<audio controls>
  <source src="path_to_your_audio.mp3" type="audio/mpeg">
Your browser does not support the audio element.
</audio>
```
Видео: Для вставки видео на веб-страницу используется тег `<video>`. Внутри этого тега размещается тег `<source>`, в атрибуте src которого указывается путь к видеофайлу. Также внутри тега `<video>` можно разместить текст, который будет отображаться в браузерах, не поддерживающих этот тег.

Пример кода:

```
<video width="320" height="240" controls>
  <source src="path_to_your_video.mp4" type="video/mp4">
Your browser does not support the video tag.
</video>
```

Все эти теги были предложены для вставки в HTML-код дочерних страниц (`dir1.html, dir2.html, dir3.html`). Важно убедиться, что пути к медиафайлам указаны правильно.

7. Решение задачи 7, где нужно добавить иконку на каждую страничку.

1. Я добавил код `<link rel="icon" href="./assets/icon.png" type="image/png">` в `<head>`.

2. `<link>` - Это HTML-тег, который используется для связи с внешними ресурсами. В данном случае он используется для связи с иконкой сайта.

3. `rel="icon"` - Атрибут rel определяет отношение между текущим документом и связанным ресурсом. В данном случае icon указывает, что связанный ресурс является иконкой сайта.

4. `href="./assets/icon.png"` - Атрибут href указывает на местоположение связанного ресурса. В данном случае он указывает на изображение icon.png в папке assets.

5. `type="image/png"` - Атрибут type указывает на MIME-тип связанного ресурса. В данном случае это image/png, что указывает на то, что иконка сайта является PNG-изображением.

1. Решение задачи 8, где нужно создать css файл и подключить его к страничкам

2. В файлике сделал таблицу стилей.
3. body и h1 - задает стили для тега body и h1.
4. `<link rel="stylesheet" href="./assets/main.css">` - подключаем файлик.
